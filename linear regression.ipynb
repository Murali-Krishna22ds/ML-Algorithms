```python
np.random.seed(42)

m=100
X = 2 * np.random.rand(m,1)
y = 4 + 3 * X +np.random.randn(m,1)
```

```python
plt.figure()
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Randomly generated Linear data")
plt.show()
```

```python
def predict(x,m,b):
    return m * x + b
```

```python
def mse(X, y, m, b):
    n=len(X)
    total=0
    for i in range(n):
        y_hat=predict(X[i], m, b)
        total+=(y[i]-y_hat)**2
    return total/n
```

```python
def gradient_descent(X, y, m, b, lr, epochs):
    n = len(X)
    
    for _ in range(epochs):
        dm = 0 
        db = 0 
        
        for i in range(n):
            y_hat = m * X[i] + b
            error = y[i] - y_hat
            dm += -2 * X[i] * error
            
        dm /= n
        db /= n 
        
        m = m - lr * dm 
        b = b - lr * db 
        
    return m, b  
```

```python
## Train the model 

m_init = 0 
b_init = 0 
lr = 0.01
epochs = 2000

m_final, b_final = gradient_descent(X, y, m_init,b_init, lr, epochs)

print("Slope (m):", m_final)
print("Intercept (b):", b_final)
```

```python
# Plot Predictions

y_pred = predict(X, m_final, b_final)

plt.scatter(X, y)
plt.plot(X, y_pred)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regression Line")
plt.show()
```
